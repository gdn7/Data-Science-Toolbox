```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
b = 1
sample = np.random.beta(a, b, 100000)

print("""
empirical mean: %.2f
analytical mean: %.2f
empirical var: %.2f
analytical var: %.2f
"""%(np.mean(sample), a / (a + b), np.var(sample), a * b / ((a + b) ** 2 * (a + b + 1))))

plt.hist(sample, bins=50)
plt.show()
```

```python
from scipy.stats import beta
a1, b1 = 10, 10
rv1 = beta(a1, b1)

a2, b2 = 3, 3
rv2 = beta(a2, b2)

x = np.linspace(0, 1, 100)
plt.plot(x, rv1.pdf(x))
plt.plot(x, rv2.pdf(x))

for spine in plt.gca().spines.values():
    spine.set_visible(False)

plt.legend(["Beta(\u03B1=%i, \u03B2=%i)"%(a1, b1), 
            "Beta(\u03B1=%i, \u03B2=%i)"%(a2, b2)], 
           frameon=False)
# plt.title("Uninformative Prior")
# plt.savefig("outputs/uninformative.png")
```

```python
from scipy.stats import beta
alphas = [0.5, 5, 1, 2, 2]
betas = [0.5, 1, 3, 2, 5]

plt.figure(figsize=(12, 6))
x = np.linspace(0, 1, 100)
for a, b in zip(alphas, betas):
    rv = beta(a, b)
    plt.plot(x, rv.pdf(x), lw=4)

plt.grid()

for spine in plt.gca().spines.values():
    spine.set_visible(False)

plt.ylim(-0.1, 2.6)
# plt.legend(["Beta(\u03B1=10, \u03B2=10)", "Beta(\u03B1=1, \u03B2=1)"], frameon=False)
# plt.title("Uninformative Prior")
plt.savefig("outputs/different_shapes_pdf.png", transparent=True)
```
