```python
df_trans = df_trans.merge(df_product, left_on="product_id", right_on="product_id")
```

```python
df_earlier = df_trans[df_trans["transaction_date"] < "2012-10-01"]
df_later = df_trans[df_trans["transaction_date"] >= "2012-10-01"]
```

```python
print(df_earlier.shape, df_later.shape)
```

```python
df_earlier["money_spent"] = df_earlier["unit_product_price"] * df_earlier["purchased_quantity"]
print(df_earlier.shape, df_later.shape)

df_earlier
```

```python
df_earlier["transaction_date"] = pd.to_datetime(df_earlier["transaction_date"], infer_datetime_format=True) 
df_earlier["days_elapsed"] = (pd.Timestamp('2012-10-01') - df_earlier["transaction_date"]) / np.timedelta64(1, 'D')
df_earlier
```

```python
df_earlier["year"] = df_earlier["transaction_date"].dt.year.astype(str)
df_earlier
```

```python
df_earlier.dtypes
```

```python
year_sum = df_earlier.groupby("year")["money_spent"].sum()
year_sum["2012"]
```

```python
df_earlier["is_red"] = df_earlier["color"] == "red"
df_earlier
```

#### Feature Engineering
* Number of transactions
* GMV (total money spent)
* Bought red?
* How many unit of red?
* Days since last order
* Money spent per transaction

```python
cust_group = df_earlier.groupby(["customer_id"])
```

```python
# Number of transactions
ntrans = cust_group.transaction_id.nunique()
ntrans
```

```python
# GMV (total money spent)
money_spent = cust_group["money_spent"].sum()
money_spent
```

```python
# Bought red?
bought_red = cust_group["is_red"].max()
bought_red
```

```python
# Money spent per transaction
avg_money_spent = cust_group["money_spent"].mean()
avg_money_spent
```

```python
# Days since most recent order
days_elapsed = cust_group["days_elapsed"].min()
days_elapsed
```

```python
# days since customer placed the first order
account_age = cust_group["days_elapsed"].max()
account_age
```

```python
df = pd.DataFrame()
df["ntrans"] = ntrans
df["money_spent"] = money_spent
df["bought_red"] = bought_red
df["avg_money_spent"] = avg_money_spent
df["days_elapsed"] = days_elapsed
df["account_age"] = account_age
df
```
